```python
import tensorflow as tf
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
import time
```

```python
base = pd.read_csv("dados_mini_dolar.csv")
base["Data"] = pd.to_datetime(base["Data"], format="%d.%m.%Y")
base = base.sort_values(by="Data",ascending=True)
base
base["Abertura"] = base["Abertura"].str.replace(".", "").str.replace(",", ".")
base=base.iloc[:,2].values
base = base.astype(np.float32).reshape(-1,1)
from sklearn.preprocessing import MinMaxScaler
scaler_base = MinMaxScaler()
base = scaler_base.fit_transform(base)

periodos = 30 
previsao_futura = 1

X_teste = base[-(periodos + previsao_futura):]
X_teste = X_teste[:periodos]
X_teste = X_teste.reshape(-1, periodos, 1)
```

```python
model = tf.keras.models.load_model("dolar.h5")
```

```python
url = "https://br.investing.com/currencies/usd-brl-mini-futures"
while True:
    response = requests.get(url)
    soup = BeautifulSoup(response.text, "html.parser")
    price = soup.find("div",{"data-test":"instrument-price-last"})
    price_att = price.text.replace(".", "").replace(",", ".")
    new_price = np.array(price_att).reshape(-1,1)
    new_price = scaler_base.transform(new_price)
    X_teste = np.delete(X_teste,0)
    X_teste = np.append(X_teste,new_price).reshape(-1,30,1)
    predict = model.predict(X_teste)
    print(scaler_base.inverse_transform(predict))
    time.sleep(1)
```
